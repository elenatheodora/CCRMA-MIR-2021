## Discussion

**What are some potential uses and benefits of hit song prediction?**


*   Record companies can determine which artists to put more resources into.
*   Songwriters and producers who have a goal of becoming commercially successful can cater their music to become hits.
*   Streaming services  can use this information to give better music recommendations to users.


**What are some potential risks or drawbacks to hit song prediction?**

*   We lose the humanness of songwriting/ production and the music listening process.
*   Record companies put resources into whatever will make money as opposed to what is artistically or musically meaningful.

Some features have values between 0 and 1, for example `danceability` and `energy`. We will call these **qualitative features**. Others have different numerical values, depending on the range of the features. These include `key` and `loudness`. We call these **quantitative features**.

## Available Features:

### Quantitative Features:
1. Artist Name and Track Name
2. Year of Release
3. Key/ Tonality - What central set of notes the song is based on
4. Loudness - Average volume of the song, measured in decibels
5. Mode - Major or minor
6. Tempo - Speed of audio in beats per minute

### Qualitative Features:
1. Danceability - Strength and regularity of the beat
2. Energy - How fast and noisy the song sounds
3. speechiness - How much spoken words are in a track
4. acousticness - Likelihood that the song uses acoustic instruments
5. instrumentalness - Detects whether a track contains no vocals
6. liveness - Detects whether the track was performed live
7. valence - How cheerful the song sounds

## Label
Finally, our data has a label. A $1$ corresponds to a 'hit' song and a $0$ corresponds to a non-hit song. This vector of labels is our "output," we don't consider it a feature.
